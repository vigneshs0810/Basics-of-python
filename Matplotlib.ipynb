```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#np.linspace(start,end,how many numbers shoulb be between equally divided)
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y,'0.8')
plt.show()
```

```python
plt.plot(x,y,'cyan')
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sinewave")
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y,'cyan')
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sinewave")
plt.show()
```

```python
fig=plt.figure()
a=fig.add_axes([0,0,1,1])
lang=["eng","spanish","latin","german","french"]
people=[100,50,150,40,70]
a.pie(people,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
a=fig2.add_axes([0,0,1,1])
a.scatter(x,y,color='g')
a.scatter(x,z,color='b')
plt.show()
```

```python
fig3=plt.figure()
a=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
a.scatter(x,y,z,c=z)
plt.show()
```
